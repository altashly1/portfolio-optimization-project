```julia
# Load the raw data
data = MyPortfolioDataSet()
original_dataset = data["train_dataset"];;
```

```julia
maximum_number_trading_days = original_dataset["AAPL"] |> nrow
```

```julia
list_of_all_tickers = keys(original_dataset) |> collect |> x->sort(x)
```

```julia
# --- Define our asset universe for this project ---
# We select a few representative tickers from your list of 424
assets_to_use = list_of_all_tickers
n = length(assets_to_use) # Number of assets

# Define the market factor
factor_ticker = "SPY"
m = 1 # Number of factors

# --- Define training and testing periods (from your notebook) ---
train_start = DateTime(2014, 1, 1)
train_end = DateTime(2023, 12, 31, 23, 59, 59)

# --- Create the training dataset for our selected tickers ---
train_data_dict = Dict{String, DataFrame}()
all_tickers_for_project = [assets_to_use; factor_ticker]

for ticker in all_tickers_for_project
    if haskey(original_dataset, ticker)
        df = original_dataset[ticker]
        df.timestamp = DateTime.(df.timestamp) # Ensure DateTime
        
        # Filter for the training period
        train_data = filter(row -> train_start <= row.timestamp <= train_end, df)
        train_data_dict[ticker] = train_data
    else
        println("Warning: Ticker $ticker not found in original_dataset.")
    end
end

println("Loaded training data for $n assets and $m factor.")
```

```julia
"""
Calculates logarithmic returns from a DataFrame's 'close' column.
"""
function calculate_log_returns(df::DataFrame)
    # Use ShiftedArrays.lag to get the previous day's close
    df.prev_close = lag(df.close, 1)
    
    # Calculate log return
    df.log_return = log.(df.close ./ df.prev_close)
    
    # Remove the first row which will have a 'missing' log_return
    return dropmissing(df, :log_return)
end

# Calculate log returns for all assets and the factor
log_returns_dict = Dict{String, Vector{Float64}}()
min_length = Inf

for (ticker, df) in train_data_dict
    returns_df = calculate_log_returns(df)
    log_returns_dict[ticker] = returns_df.log_return
    if length(returns_df.log_return) < min_length
        min_length = length(returns_df.log_return)
    end
end

# Ensure all return vectors are the same length by truncating to the minimum
# This aligns all the dates
p = Int(min_length) # Number of observations
println("Number of trading days (observations) p = $p")

log_returns_df = DataFrame(timestamp = train_data_dict["SPY"].timestamp[end-p+1:end])
for ticker in all_tickers_for_project
    full_vec = log_returns_dict[ticker]
    log_returns_df[!, ticker] = full_vec[end-p+1:end]
end

display(first(log_returns_df, 5))
```

```julia
# --- 1. Set up regression inputs ---
f = log_returns_df[!, factor_ticker] # Factor returns
A = [ones(p) f]                      # Design matrix A = [1 B^T] [cite: 462]

# --- 2. Set Confidence Level ---
# We choose ω = 0.95, a common choice in the paper's tests [cite: 1029]
ω = 0.95
# Critical value from F-distribution [cite: 521, 535]
c_J = quantile(FDist(m, p-m-1), ω) 
println("Confidence level ω = $ω, Critical F-value c_J = $c_J")

# --- 3. Pre-calculate regression-related matrices ---
# (AᵀA)⁻¹ is needed for uncertainty bounds [cite: 471]
A_inv = inv(A'A)
A_inv_11 = A_inv[1, 1] # For γ (mean uncertainty)
# G = (Q(AᵀA)⁻¹Qᵀ)⁻¹[cite: 540], for m=1, Q=[0 1], this simplifies
# We only need the component for the factor, not the full matrix G
# The paper's formulation for ρ in (63) is what we follow
G_inv_component = A_inv[2,2] 

# --- 4. Initialize parameter vectors ---
μ₀ = Vector{Float64}(undef, n)
V₀ = Vector{Float64}(undef, n)
γ = Vector{Float64}(undef, n)
ρ = Vector{Float64}(undef, n)
D_bar = Vector{Float64}(undef, n)

# --- 5. Run regression for each asset ---
println("Running linear regression for each asset...")
for (i, ticker) in enumerate(assets_to_use)
    y = log_returns_df[!, ticker] # Asset returns
    
    # Run the linear regression: y = A*x + ϵ [cite: 461]
    model = lm(A, y)
    
    # Get estimates for μ₀ and V₀
    params = coef(model)
    μ₀[i] = params[1]  # Intercept
    V₀[i] = params[2]  # Slope for "SPY" factor
    
    # Get residual variance sᵢ²
    # sᵢ² = ||y - Ax̄||² / (p-m-1) 
    # This is the deviance divided by the residual degrees of freedom
    s_i_squared = deviance(model) / dof_residual(model)
    D_bar[i] = s_i_squared
    
    # Calculate uncertainty bounds per (62) and (63)
    # γᵢ = sqrt((AᵀA)⁻¹₁₁ * c₁(ω) * sᵢ²) [cite: 62]
    γ[i] = sqrt(A_inv_11 * quantile(FDist(1, p-m-1), ω) * s_i_squared) 
    
    # ρᵢ = sqrt(m * cₘ(ω) * sᵢ²) [cite: 63] (m=1 here)
    ρ[i] = sqrt(m * quantile(FDist(m, p-m-1), ω) * s_i_squared)
end

println("Parameter estimation complete.")

# Display estimated parameters
param_df = DataFrame(
    Ticker = assets_to_use,
    μ₀ = μ₀,
    γ_uncertainty = γ,
    V₀_loading = V₀,
    ρ_uncertainty = ρ,
    ResidualVar_D_bar = D_bar
)
display(param_df)
```

```julia
# Factor covariance matrix F (since m=1, this is just a scalar)
F = var(f)
println("Factor Variance F = $F")

# Kappa parameter for the simplification F = κG
# p-1 is used for sample variance, which matches G's derivation 
κ = 1.0 / (p - 1)
println("Model simplification parameter κ = $κ")

# Risk-free rate (r_f)
# The paper's real-data example sets r_f = 0 [cite: 1020]
r_f_daily = 0.0
```

```julia
# Initialize the JuMP model
model = Model(SCS.Optimizer)
set_silent(model) # Hide solver output

# --- Define Optimization Variables ---
@variable(model, ϕ[1:n] >= 0) # Portfolio weights (non-negative)
@variable(model, ν >= 0)      # Slack for factor variance
@variable(model, δ >= 0)      # Slack for residual variance

# --- Set Objective ---
@objective(model, Min, ν + δ)

# --- Define Constraints ---

# Constraint 1: Worst-case excess return 
# (μ₀ - γ - r_f_daily * ones(n))' * ϕ >= 1
@constraint(model, return_constraint, (μ₀ - γ)' * ϕ >= 1)

# Constraint 2: Residual variance [cite: 359]
# This is the SOCP form of: ϕ' * Diagonal(D_bar) * ϕ <= δ
# which is || [2*sqrt(D_bar)ϕ; 1-δ] ||₂ <= 1+δ
@constraint(model, residual_variance, 
    [1 + δ; 2 * sqrt.(D_bar) .* ϕ; 1 - δ] in SecondOrderCone()
)

# Constraint 3: Factor variance (simplified model) 
# We linearize the absolute value in: sqrt(F) * |V₀' * ϕ| + sqrt(κ) * (ρ' * ϕ) <= ν
factor_exposure = V₀' * ϕ
@constraint(model, factor_variance_pos, 
    sqrt(F) * factor_exposure + sqrt(κ) * (ρ' * ϕ) <= ν
)
@constraint(model, factor_variance_neg, 
    -sqrt(F) * factor_exposure + sqrt(κ) * (ρ' * ϕ) <= ν
)

# --- Solve the Optimization Problem ---
optimize!(model)

println("Optimization finished.")
println("Solver status: $(termination_status(model))")
```

```julia
if termination_status(model) == OPTIMAL
    # Get the unnormalized weights
    ϕ_solution = value.(ϕ)
    
    # Normalize the weights to get the final portfolio
    ϕ_normalized = ϕ_solution / sum(ϕ_solution)
    
    println("\n--- Optimal Robust Portfolio Allocation ---")
    
    results_df = DataFrame(
        Ticker = assets_to_use,
        Weight = ϕ_normalized
    )
    
    # Sort by weight and display
    display(sort(results_df, :Weight, rev=true))
    
    println("\nTotal Allocation: $(sum(ϕ_normalized))")
else
    println("Error: The model did not find an optimal solution.")
end
```
